## Benchmark 1: 1D Riemann problems

### Simulation parameters


Libraries are imported and program folders and files are defined:

```python



import math   

import numpy as np             
import matplotlib.pyplot as plt 

folder_caseI="caseI"
folder_out="outputI-files"

fname_config="configureI.input"
fname_ini="initial.out"

#data must be deleted before running
```

Simulation configuration:

```python
#Simulation setup
Beta_inicial = 0.0
Beta_final = 2.0
Delta_beta = 0.001
beta_analizar=1
Semilla =12345
#Mesh setup
N_ter = 200
N_medidas = 1000
N_bloques = 5
N_cambios=10
#condiciones inciales
modo_ejecucion = 0
#DIMENSIONES
L = 128

n_beta=int((Beta_final-Beta_inicial)/Delta_beta)


```

### Initial conditions

### Writing files

```python
f = open(folder_caseI+"/"+fname_config, "w")

f.write("FinalTime    "+str(L)+"\n")
f.write("DumpTime    "+str(Semilla)+"\n")
f.write("CFL    "+str(Beta_inicial)+"\n")
f.write("Order    "+str(Beta_final)+"\n")
f.write("xcells    "+str(Delta_beta)+"\n")
f.write("ycells    "+str(N_ter)+"\n")
f.write("zcells    "+str(N_cambios)+"\n")
f.write("SizeX    "+str(N_medidas)+"\n")
f.write("SizeY    "+str(N_bloques)+"\n")
f.write("SizeZ    "+str(modo_ejecucion)+"\n")
f.write("Face_1    "+str(beta_analizar)+"\n")
f.close()


       
```

### Running the code

```python
import subprocess

binary="./ISING"      

def system_call(command):
    p = subprocess.Popen([command], stdout=subprocess.PIPE, shell=True)
    while True:
        line = p.stdout.readline()
        if not line:
            break
        print(line.rstrip())
    return 1

system_call(binary)
```

### Reading and plotting data:

Reading data:

```python
from glob import glob

files = glob(folder_out+"/list*")
lf=len(files)




beta=np.zeros((n_beta))
medida_E=np.zeros((n_beta))
error_e=np.zeros((n_beta))
medida_m=np.zeros((n_beta))
error_m=np.zeros((n_beta))
med_cuad_E=np.zeros((n_beta))
med_cuad_m=np.zeros((n_beta))
print(n_beta)
for fname in files:
    data     = np.loadtxt(fname)
    j=0

n=0
for m in range(0,n_beta):
        beta[m]=data[n,0]
        medida_E[m]=data[n,1]
        error_e[m]=data[n,2]
        medida_m[m]=data[n,3]
        error_m[m]=data[n,4]
        med_cuad_E[m]=data[n,5]
        med_cuad_m[m]=data[n,6] 
        n=n+1          
                      
    
print(beta[:])
print(medida_E[:])
print(error_e[:])
print(medida_m[:])
print(error_m[:])
print(med_cuad_E[m])
print(med_cuad_m[m])
```

**1D PLOT:**

```python
j=1

filename = "doc/bench_RP_"

fig, ax  = plt.subplots(2,2,figsize=(14, 8))

ax[0,0].plot(beta[:],medida_m[:],'o-')

ax[0,0].set_xlabel("beta") 
ax[0,0].set_ylabel("magnetización") 


ax[1,0].plot(beta[:],medida_E[:],'o-')

ax[1,0].set_xlabel("beta") 
ax[1,0].set_ylabel("energía") 

#print(np.max(phi))
#print(np.min(phi))

fig.savefig(filename+".png",dpi=500)
```
